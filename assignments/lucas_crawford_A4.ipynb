```python
class Soldier:
    def __init__(self, value):
        self.value = value
        self.next = None
        self.prev = None
    
    def kill(self, prefix, suffix):
        print(prefix + str(self.value) + suffix)

class Army:
    def __init__(self, n):
        self.soldierCount = n
        self.head = None
        self.currentVictim = None
        for i in range(1, n+1):
            self.addSoldier(i)
        self.currentVictim = self.head
    
    def addSoldier(self, i):
        soldier = Soldier(i)
        if self.head is None:
            self.head = soldier
            self.head.next = self.head # in this case just set the next reference to itself
            self.head.prev = self.head # same as above for prev
        else:
            # insert a new soldier at 'end' of DLL
            lastSoldier = self.head.prev
            
            soldier.prev = lastSoldier
            soldier.next = self.head
            lastSoldier.next = soldier
            self.head.prev = soldier
    
    def advance(self, k):
        while k-1 != 0:
            self.currentVictim = self.currentVictim.next
            k -= 1

    def removeSoldier(self, soldier):
        soldier.prev.next = soldier.next
        soldier.next.prev = soldier.prev
        self.soldierCount -= 1
        if self.head == soldier:
            self.head = self.head.next
        if self.soldierCount > 0:
            self.currentVictim = self.currentVictim.next

    def killAll(self, k):
        while self.soldierCount > 1:
            self.advance(k)
            self.currentVictim.kill('Soldier ', ' died by their own hand.')
            self.removeSoldier(self.currentVictim)
        print('The last remaining soldier is', self.currentVictim.value)
```

```python
n = int(input("Enter number n of soliders, at least 2: "))
k = int(input("Enter spacing between victims, between 1 and n: "))

army = Army(n)
army.killAll(k)
```
